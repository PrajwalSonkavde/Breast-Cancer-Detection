```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df['State'].unique()
```

```python
plt.figure(figsize=(30,17))
ax = plt.axes()
ax.set_facecolor("#2a2727")
plt.plot(df['State'],df['Rate'],marker="*",markersize=15,linewidth=5)
plt.xlabel('State')
plt.ylabel("Rate")
plt.title("People died rate due to cancer in US (State wise )")
plt.savefig("graph.png")

plt.show()
```
